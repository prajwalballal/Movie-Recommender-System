## Part 1: Content based recommendation system

### *Here We are finding 50 similar movies that user has entered based on keywords for the movies, Genre of the movie,Director of the movie and the cast using Cosine similarity as a measure of similarity.*

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# To Get the index of the movie that matches the title
def get_t_from_i(index):
    return df[df.index == index]['title'].values[0]

def get_i_from_t(title):
    return df[df.title == title]['index'].values[0]
```

```python
df= pd.read_csv("D:\\Praxis\\Material\\Term 2\\ML\\Recommendation system\\movie_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
features= ['keywords','cast','genres','director']
```

```python
for i in features:
    df[i] = df[i].fillna("")
```

```python
def combine_features(row):
    return row['keywords'] +" " +row['cast']+" "+row['genres']+" "+row['director']
```

```python
df['combine_features'] = df.apply(combine_features,axis=1)
```

```python
cv = CountVectorizer()

count_matrix = cv.fit_transform(df['combine_features'])

cos_sim = cosine_similarity(count_matrix)
```

```python
movie_user_likes = input("ENTER A MOVIE YOU WOULD LIKE RECOMMENDATION FOR ? ==>")
    
movie_index = get_i_from_t(movie_user_likes)

#Get the pairwsie similarity scores of all movies with that movie
similar_movies = list(enumerate(cos_sim[movie_index]))
# Sort the movies based on the similarity scores
sorted_similar_movies = sorted(similar_movies,key= lambda x:x[1],reverse= True)

# Return the top 50 most similar movies
i=0
for movie in sorted_similar_movies:
    print(get_t_from_i(movie[0]))
    i= i+1
    if i>50:
        break
```

## Part 2: Using Nearest Neighbour clustering

```python
def recommender(t):
    import numpy as np
    import pandas as pd
    import os 
    
    # reading the data
    # you need to change the directory according to your database location
    data = pd.read_csv("D:\\Praxis\\Material\\Term 2\\ML\\Recommendation system\\movies.csv")
    # dropping the ID columns as we dont need it 
    data = data.drop('movieId',axis=1)
    
    # making the title of movies as index and sepearting the genres and making different
    # columns for each unique genre and making a dense matrix
    newt1 = data.set_index(['title']).genres.str.split('|', expand = True).stack()
    final_movies = pd.get_dummies(newt1).groupby(level =0 ).sum()
    
    # using csr_martrix function to convert above dense matrix into sparse matrix
    from scipy.sparse import csr_matrix
    movie_features_df_matrix = csr_matrix(final_movies.values)

    # using Nearest Neighbour to make clusters based on genres
    from sklearn.neighbors import NearestNeighbors
    model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute') # using cosine similarity as distance metric
    model_knn.fit(movie_features_df_matrix)
    
    # this is for finding the index number of the movie entered by the user
    for position, item in enumerate(final_movies.index):
        if item == t:
            query_index = position # this contains the index
            
      
    # storing the distance and indexes of the nearest movies to the entered movie
    distances, indices = model_knn.kneighbors(final_movies.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 11)
    
    # for printing the top 10 similar movies to the entered movie
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Recommendations for {0}:\n'.format(final_movies.index[query_index])) # beacuse the first movie will always be the movie entered
        else:
            print('{0}: {1}'.format(i, final_movies.index[indices.flatten()[i]]))
```

```python
recommender('Tom and Huck (1995)')  # need to give movie name in string format
```

## Pearson similarity 

```python
#We first fit the recommendation engine with Pearson similairty technique.
item_simlilarity_df= user_rating_threshold.corr(method='pearson')

item_simlilarity_df.head(10)
```

### *Below algorithm takes two arguments, viz. Name of movie & Rating of user for the same movie.*

*Interesting thing about below code is, we can set threshold for rating, below which movie is considered to be bad. So if user rates movie lesser than threshold set, machine has to recommend a movie which is totally different in nature than what he watched.*

*Logic is, we multiply whole series corresponding to movie by a negative number if rating is lesser than threshold. Hence we get recomendation which are different than the movie rated. If the rating of movie is more than the threshold then we recomend the similar movies.*

```python
def get_similar_movies_pearson(movies_name, rating):
    similar_score= item_simlilarity_df[movies_name]*(rating-2.5)
    similar_score= similar_score.sort_values(ascending=False)
    return similar_score.head(10)
```

```python
#Keeping the list of movies watched and respective rating in tuples. 
user= [('Toy Story (1995)',1),('127 Hours (2010)',4.5)]
        
similar_movie= pd.DataFrame()

for movie, rating in user:
    similar_movie= similar_movie.append(get_similar_movies_pearson(movie, rating), ignore_index=True)
    
# Recommending top 10 movies.

similar_movie.head(10)
similar_movie.sum().sort_values(ascending=False).head(10)
```

### Below function does not take rating as an argument. Only takes name of movie. Recomends user based on similarity of movies (Item to Item similarity).

```python
def get_similar_test(movies_name):
    similar_score= item_simlilarity_df[movies_name]
    similar_score= similar_score.sort_values(ascending=False)
    return similar_score
```

### Since we have replaced NA values with 0, we need to standardize all ratings to make sure that there is no bias. 

### New_Rating= (Rating- Mean_Rating)/(Max.Rating-Min.Rating)
### It does mean of each row =0, and range =1.

```python
def standardize(row):
    new_row= (row- row.mean() )/ (row.max()- row.min())
    return new_row
```

```python
data_pivot_STD= data_pivot_2.apply(standardize, axis=1)
```

```python
item_similarity= cosine_similarity(data_pivot_STD.T)
```

```python
item_similarity_df= pd.DataFrame(item_similarity, index= data_pivot_STD.columns, columns=data_pivot_STD.columns )
```

```python
# Lets create recomendation 

def get_sim_movie_cosine(movie, rating):
    similar_score= item_similarity_df[movie]*rating
    similar_score= similar_score.sort_values(ascending=False)
    return similar_score

"""This method works fine if only user likes the movie he entered. What if user does not like a movie?
Then we dont want to suggest same kind of movies he watched. We need to set threshold value, below which we consider as bad movie.
We mupltiply the consine similarity scores by difference factor, so that we get appropriate movie recomendation
    
"""
```
